Criando uma analise para desenvolver um Dashboard sobre as vendas de uma empresa


```python
#Importando as bibliotecas para o código
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
#Carregando os dados da Base CSV

tabela = pd.read_csv("vendas.csv",sep=";",encoding="latin1")
display(tabela)

#Ajustando os tipos de dados 
tabela["Data"] = pd.to_datetime(tabela["Data"]) #Convertando as datas para Ano/Mes/ Dia
tabela["Receita"] = tabela["Quantidade"] * tabela["Valor"] #Criando uma Coluna nova para verificar a receita do produtos (Quantidade * Valor)
display(tabela)
```

Criando a Analise da Tabela 
    -Receita Total
    -Vendas por Categoria
    -Produto Mais Vendido
    -Receita do Mês

```python
#Receita Total

receita_total = tabela["Receita"].sum()
print(receita_total)
```

```python
#Vendas por categoria
tabela = tabela.replace("AcessÃ³rios","Acessorios").replace("MÃ³veis","Moveis") #Renomeando os Valores das Cedulas(AcessÃ³rios ,MÃ³veis)
vendas_categoria = tabela.groupby("Categoria")["Quantidade"].sum() #Separando as Categorias

print(vendas_categoria)
```
