Criando uma analise para desenvolver um Dashboard sobre as vendas de uma empresa


```python
#Importando as bibliotecas para o código
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
```

```python
#Carregando os dados da Base CSV

tabela = pd.read_csv("vendas.csv",sep=";",encoding="latin1")
display(tabela)

#Ajustando os tipos de dados 
tabela["Data"] = pd.to_datetime(tabela["Data"]) #Convertando as datas para Ano/Mes/ Dia
tabela["Receita"] = tabela["Quantidade"] * tabela["Valor"] #Criando uma Coluna nova para verificar a receita do produtos (Quantidade * Valor)
display(tabela)
```
