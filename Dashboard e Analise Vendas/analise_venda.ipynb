Criando uma analise para desenvolver um Dashboard sobre as vendas de uma empresa


```python
#Importando as bibliotecas para o código
import pandas as pd
import plotly.express as px
import plotly.graph_objects as go
import plotly.io as pio
```

```python
#Carregando os dados da Base CSV

tabela = pd.read_csv("vendas.csv",sep=";",encoding="latin1")
display(tabela)

#Ajustando os tipos de dados 
tabela["Data"] = pd.to_datetime(tabela["Data"]) #Convertando as datas para Ano/Mes/ Dia
tabela["Receita"] = tabela["Quantidade"] * tabela["Valor"] #Criando uma Coluna nova para verificar a receita do produtos (Quantidade * Valor)
display(tabela)
```

Criando a Analise da Tabela 
    -Receita Total
    -Vendas por Categoria
    -Produto Mais Vendido
    -Receita do Mês

```python
#Receita Total

receita_total = tabela["Receita"].sum()
print(receita_total)
```

```python
#Vendas por categoria
tabela = tabela.replace("AcessÃ³rios","Acessorios").replace("MÃ³veis","Moveis") #Renomeando os Valores das Cedulas(AcessÃ³rios ,MÃ³veis)
vendas_categoria = tabela.groupby("Categoria")["Receita"].sum().reset_index() #Separando as Categorias

print(vendas_categoria)
```

```python
#Produto Mais Vendido
produto_mais_vendido = tabela.groupby("Produto")["Quantidade"].sum().sort_values(ascending=False).head(10)
print(produto_mais_vendido)
```

```python
#Receita por Mês
tabela["Mes"] = tabela["Data"].dt.to_period("M") #Cria uma Linha Mes a partir da Coluna Data
receita_por_mes = tabela.groupby("Mes")["Receita"].sum().reset_index() #Agrupa todos os registros pelos Mes
print(receita_por_mes)
```

Criar Gráficos Interativos para Visualização:
    -Gráfico de Receita por Categoria
    -Gráfico Top 10 Produtos Vendidos
    -Gráfico Evolução da Receita Mensal

```python
#Gráfico de Receita por Categoria
grafico_receita_categoria = px.bar(
    vendas_categoria,
    x = "Categoria",
    y = "Receita",
    title= "Gráfico de Receita por Categoria",
    text_auto=True
    
)
grafico_receita_categoria.show()
```

```python
#Gráfico Top 10 Produtos Vendidos
grafico_produtos = px.bar(
    produto_mais_vendido,
    x =produto_mais_vendido.index,
    y= produto_mais_vendido.values,
    title="Gráfico Top 10 de Produtos mais Vendidos",
    text_auto=True
)
grafico_produtos.show()
```

```python
#Gráfico Evolução da Receita Mensal
grafico_receita_mensal = px.line(
    receita_por_mes.astype({"Mes": str}),
    x="Mes",
    y="Receita",
    markers=True,
    title=f"Evolução da Receita por Mês - Periodo {receita_por_mes}"
)
grafico_receita_mensal.show()
```
