```python
import pandas as pd #Importando o Padas
import plotly.express as px
```

```python
#Importando a base de dados 

tabela = pd.read_csv("ClientesBanco.csv", encoding="latin1")
display(tabela)
```

```python
#Tratando a base de Dados
#Excluindo a colunas colunas da Tabela
tabela = tabela.drop("CLIENTNUM", axis=1)
display(tabela)
```

```python
#Visualizando os dados das colunas verificando o tipo de dados 

tabela = tabela.dropna()#Excluindo todas as linhas qaundo em um coluna tem pelo menos 1 valor vazio
display(tabela.info()) #Verificando se tem informações vazias

display(tabela.describe().round(1)) #Metodo para descrever os dados da Tabela 
```

Construindo os Gráficos, trazendo as analizes
- Fazer comparação em colunas 

```python
#Passando informações para o Grafico - Grafico da Idadde e Categoria
#Tabela
#Eixo X
#Cor - Referece a comparação de filtro, aonde vou passar mais uma coluna
grafico = px.histogram(tabela, x="Idade", color="Categoria")
grafico.show() #Exibe o Gráfico
```

```python
#Grafico - Categoria / Educação
grafico = px.histogram(tabela, x="Categoria",color="Educação")
grafico.show()                    
```

Trabalhando FOR dentro da Tabela
- O For dentro do Pyhton não percorrer as linhas Percorree as colunas
- Para percorrer as linhas precisa adicionar um código na frente da tabla "for linha in tabela.index"

```python
#Fazendo Filtros dentro das Colunas
#Verificando se existe valores nas colunas diferentes
#Utilizando a coluna Sexo , os Dados é F e M - Verificando se existe algum valor diferente

coluna_sexo_1 = tabela[(tabela["Sexo"] != "M") & (tabela["Sexo"] != "F")] # primeira forma de comparar dois valores dentro do IF
coluna_sexo_2 = tabela[~tabela["Sexo"].isin(["M","F"])]
print(len(coluna_sexo_1))
print(len(coluna_sexo_2))
```

```python
## Brincando com as análises ##
#Exibindo o gráfico para todas as colunas 
#Primeira Análise - Filtro em Cima de todos os clientes de Sexo Masculino
#Realizei o Comparativo da Categoria(Cliente / Cancelado)
#Relizei a cor do gráfico trazendo a Categoria do Cartão (Blue,Gold,Silver,Platinum) 
#Criei um gráfico de Histograma

coluna_sexo = tabela[tabela["Sexo"] == "M"]
total_m = len(coluna_sexo)


grafico = px.histogram(
        coluna_sexo,
        x="Categoria",
        color="Categoria Cartão",
        title=f"Gráfico de Clientes e Cancelados pelo Sexo Masculino - Dados: {total_m}"
)

grafico.show()
```

```python
#Primeira Análise - Filtro em Cima de todos os clientes de Sexo Feminino
#Realizei o Comparativo da Categoria(Cliente / Cancelado)
#Relizei a cor do gráfico trazendo a Categoria do Cartão (Blue,Gold,Silver,Platinum) 
#Criei um gráfico de Histograma

coluna_sexo = tabela[tabela["Sexo"]=="F"]
total_f = len(coluna_sexo)
print(total_f)

grafico= px.histogram(
            coluna_sexo,
            x = "Categoria", 
            color= "Categoria Cartão",
            title=f"Gráfico de Clientes e Cancelados pelo Sexo Feminino - Dados: {total_f}")
grafico.show()
```

```python
#Analisando Os  Dependetes

coluna_dependentes=tabela[tabela["Dependentes"]>=3]
total_dependentes = len(coluna_dependentes)

grafico = px.histogram(
    coluna_dependentes,
    x = "Categoria",
    color='Categoria Cartão',
    title= f"Gráfico de Clientes e Cancelados referente ao número de Dependentes Maior que 3 - Dados: {total_dependentes}"
)
grafico.show()
```
